```python
import numpy as np
```

```python
arr=np.random.randint(low=1,high=100,size=20).reshape(4,5)

arr
```

```python
#indexing using boolen array  and filtering 
arr[0][arr[0]>50]
```

```python
arr[(arr>50) & (arr%2!=0)]
```

```python
arr[(arr>50) & (arr%2!=0)]=0
```
